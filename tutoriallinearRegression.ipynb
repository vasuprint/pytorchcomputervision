```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
class LR(nn.Module):
    def __init__(self,input_size, output_size):
        super().__init__()
        self.linear = nn.Linear(input_size, output_size)        
    def forward(self,x):
        pred = self.linear(x)
        return pred
    
    def get_params(self):
        self.w, self.b = self.parameters()
        return w[0][0].item(), b[0].item()
    
    def plot_fit(self, title):
        plt.title = title
        self.w1, self.b1 = self.get_params()
        x1 = np.array([-30,30])
        y1 = self.w1*x1 + self.b1
        plt.plot(x1,y1,'r')
        plt.scatter(X,y)
        plt.show()
```

```python
X = torch.randn(100,1)*10
y = X + 3*torch.randn(100,1)
model = LR(1,1)
[w,b] = model.parameters()
```

```python
plt.scatter(X,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
model.plot_fit('Initial Model')
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
losses = []
epochs = 100
for i in range(epochs):
    y_pred = model.forward(X)
    loss = criterion(y_pred,y)
    print('epoch:', i, 'loss:', loss.item())
    
    losses.append(loss.item())
    optimizer.zero_grad() # to reset the gradients because pytorch accumulates the gradients it helpful to use with RNN
    loss.backward()
    optimizer.step()
```

```python
plt.plot(range(epochs), losses)
```
